```python
influence=pd.read_csv('influence.csv')
df=pd.DataFrame(influence)
column_name='Your grade in a course, where you had a good (positive impactful) instructor'
column_data=df[column_name]
```

```python
value_counts=column_data.value_counts()
plt.figure(figsize=(10, 7))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(column_name)
plt.axis('equal')
plt.show()
```

```python
column_name2='Your grade in a course, where you had a poor (negative impactful) instructor'
column_data2=df[column_name2]
value_counts=column_data2.value_counts()
plt.figure(figsize=(10, 7))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title(column_name2)
plt.axis('equal')
plt.show()
```

```python
df = pd.read_csv('influence.csv')

column_name = 'Your grade in a course, where you had a good (positive impactful) instructor'

plt.figure(figsize=(10, 6))
plt.bar(df[column_name].value_counts().index, df[column_name].value_counts())
plt.title('Bar Chart of ' + column_name)
plt.xlabel('Grades')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
